# 自分の解答

```python
a,b=map(int,input().split())
 
if (a > b):
    print("safe")
else:
    print("unsafe")
```

# B問題

# 自分の解答

```python
a,b,c,d=map(int,input().split())
 
x = c/b
y = a/d
 
if x > int(x):
    x+=1
    x=int(x)
if y > int(y):
    y+=1
    y=int(y)

if x <= y:
    print("Yes")
else:
    print("No")
```

# 自分の解答

```python
a = int(input())
lst = []
for i in range(a):
    b = input()
    if not(b in lst):
        lst.append(b)
        
print(len(lst))
```
